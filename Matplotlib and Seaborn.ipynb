```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.DataFrame(np.random.randint(1,100,(4,3)),index=['Riya','Siya','Kiya','Jiya'],columns=['python','DSA','Data_Science'])
```

```python
df
```

```python
f,ax=plt.subplots(1,3)
```

```python
ax[0].plot(df.index,df.python,'g',linestyle='-.')
ax[0].set_title('Python')

ax[1].plot(df.index,df.DSA,'rd:')
ax[1].set_title('DSA')

ax[2].plot(df.index,df.Data_Science,'k-s')

ax[2].set_title('DATA Science')
```

```python
f
```

```python
fig,axes = plt.subplots(2,2,layout='constrained',sharex=True,sharey=True)
```

```python
axes[1,1].plot(df.index,df.DSA,'gs:',label='DSA')
axes[1,1].set_xlabel('name')
axes[1,1].set_ylabel('dsa marks')
axes[1,1].set_title('DSA')
axes[1,1].legend()


axes[0,0].plot(df.index,df.python,'r',label='Py')
axes[0,0].set_xlabel('name')
axes[0,0].set_ylabel('python marks')
axes[0,0].set_title('Python')
axes[0,0].legend()

axes[0,1].plot(df.index,df.Data_Science,'kd',label='Data Science')
axes[0,1].set_xlabel('name')
axes[0,1].set_ylabel('Data Science marks')
axes[0,1].set_title('Data Science')
axes[0,1].legend()

axes[1,0].plot(df.index,df.Data_Science,df.python,df.DSA,'pink',linewidth=2)
axes[1,0].set_xlabel('name')
axes[1,0].set_ylabel('ALL marks')
axes[1,0].set_title('ALL')
axes[1,0].legend()
```

```python
import seaborn as sns
%matplotlib inline
```

```python
import numpy as np
h=np.random.randint(0,100,12).reshape(4,3)
v=["dsa","stb","cmp"]
h
```

```python
import pandas as pd
df=pd.DataFrame(h,index=["Riya","Rahul","Ramesh","Rohit"],columns=v)
df
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.displot(df['dsa'],kde=False,bins=15)
```

```python
sns.jointplot(x='dsa',y='cmp',data=df,kind='scatter')
```

```python
sns.pairplot(df,palette='brg')
```

```python
sns.rugplot(df['dsa'])
```

```python
sns.heatmap(df.corr())
```
